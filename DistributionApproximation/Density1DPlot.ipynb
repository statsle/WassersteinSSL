```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
import os
import numpy as np

plt.rcParams['axes.linewidth'] = 1  # 
plt.rcParams['figure.dpi'] = 300
plt.rc('font',family='Times New Roman')

figsize=(12, 7.2)
tick_fontsize = 30
label_fontsize  = 35

data = loadmat("probs-2.mat")
value = (np.array(data["value"])+ 1/51).tolist()[0]
source_density = data["source_density"].tolist()[0]
target_density = data["target_density"].tolist()[0]
source_density = [100*v for v in source_density]
target_density = [100*v for v in target_density]

fig = plt.figure(figsize=figsize)
plt.xlim((-1.01, 1.02))
x_ticks = [-1.0, -0.5, 0.0, 0.5, 1.0]
x_labels = [-1.0, -0.5, 0.0, 0.5, 1.0]
plt.xticks(x_ticks, x_labels, fontsize=tick_fontsize)

plt.ylim((-0.05, 10))
y_ticks = [0, 2, 4, 6, 8, 10]
y_labels = [0, 2, 4, 6, 8, 10]
plt.yticks(y_ticks, y_labels, fontsize=tick_fontsize)

plt.plot(value, source_density, 'o', color='blue',  linewidth=2, markersize=8, linestyle="-.")
plt.plot(value, target_density, 'P', color='green', linewidth=2, markersize=8, linestyle="--")

plt.legend(labels=[ ''r'$Y_i$', ''r'$\hat{Y}_i$'], loc=1, fontsize=tick_fontsize)
plt.gca().yaxis.grid(color='grey', linestyle='-', linewidth=1, alpha=0.3, zorder=0)
plt.gca().xaxis.grid(color='grey', linestyle='-', linewidth=1, alpha=0.3, zorder=0)
plt.xlabel('Variable', fontsize=label_fontsize)
plt.ylabel('Binning Density (x 0.01)', fontsize=label_fontsize)
plt.savefig('Binning_Density-2.pdf', bbox_inches='tight')
```
