```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import os

plt.rcParams['axes.linewidth'] = 1  # 图框宽度
plt.rcParams['figure.dpi'] = 300
plt.rc('font',family='Times New Roman')

figsize=(12, 7.2)
tick_fontsize=20
label_fontsize  = 15

data = loadmat("probs2D.mat")
value = (np.array(data["value"])+ 1/51).tolist()[0]
source_density = np.array(data["source_density"].tolist())
target_density = np.array(data["target_density"].tolist())

X, Y = np.meshgrid(value, value)

fig = plt.figure(figsize=figsize)
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, source_density, rstride=1, cstride=1, cmap=plt.cm.coolwarm, edgecolor='none')

#ax.plot_surface(X, Y, source_density, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
#ax.plot_surface(X, Y, source_density, rstride=1, cstride=1, color='blue', edgecolor='none')
ax.set_xlabel(''r'$Y_i$',fontsize=label_fontsize)
ax.set_ylabel(''r'$Y_j$',fontsize=label_fontsize)
ax.set_zlabel('Binning Density',fontsize=label_fontsize)
plt.savefig('Binning_Density_2D_source.pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=figsize)
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, target_density, rstride=1, cstride=1, cmap=plt.cm.coolwarm, edgecolor='none')
ax.set_xlabel(''r'$\hat{Y}_i$',fontsize=label_fontsize)
ax.set_ylabel(''r'$\hat{Y}_j$',fontsize=label_fontsize)
ax.set_zlabel('Binning Density',fontsize=label_fontsize)
plt.savefig('Binning_Density_2D_target.pdf', bbox_inches='tight')
```
