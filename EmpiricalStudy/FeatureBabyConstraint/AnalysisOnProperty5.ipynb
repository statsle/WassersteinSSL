```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import os
import seaborn as sns
import pandas as pd
from scipy.io import loadmat

load_data = loadmat("property5.mat")
times = load_data["times"]
w2_results = load_data["w2_results"]
lu_results = load_data["lu_results"]

print(times)
print(w2_results)
print(lu_results)

plt.rcParams['axes.linewidth'] = 1  # 图框宽度
plt.rcParams['figure.dpi'] = 300
plt.rc('font',family='Times New Roman')

figsize=(12, 7.2)
tick_fontsize = 30
label_fontsize  = 35

fig = plt.figure(figsize=figsize)

plt.xlim((-0.1, 10.1))

x_ticks =  [0, 2, 4, 6, 8, 10]
x_labels = [0, 2, 4, 6, 8, 10]
plt.xticks(x_ticks, x_labels, fontsize=tick_fontsize)

plt.ylim((-1.5, 0))

y_ticks =  [-1.5, -1.2, -0.9, -0.6, -0.3, 0]
y_labels = [-1.5, -1.2, -0.9, -0.6, -0.3, 0]
plt.yticks(y_ticks, y_labels, fontsize=tick_fontsize)
plt.gca().yaxis.grid(color='grey', linestyle='-', linewidth=1, alpha=0.3, zorder=0)
plt.gca().xaxis.grid(color='grey', linestyle='-', linewidth=1, alpha=0.3, zorder=0)

plt.plot(times[0], w2_results[0]*(-1.0), 'P', color='blue', linewidth=2, markersize=12, linestyle="-.")
plt.plot(times[1], w2_results[1]*(-1.0), 'd', color='green', linewidth=2, markersize=12, linestyle="--")
plt.plot(times[2], w2_results[2]*(-1.0), 'H', color='coral', linewidth=2, markersize=12, linestyle="-")

plt.legend(labels=[ ''r'$\eta=75\%$', ''r'$\eta=50\%$',''r'$\eta=25\%$',], loc=1, fontsize=tick_fontsize)

plt.xlabel('Dimensions of Feature Baby ($i=k/m$)', fontsize=label_fontsize)
plt.ylabel('$-\mathcal{W}_\mathcal{2}$', fontsize=label_fontsize)
plt.savefig('AnalysisOnFBC_W2.pdf', bbox_inches='tight')
```
