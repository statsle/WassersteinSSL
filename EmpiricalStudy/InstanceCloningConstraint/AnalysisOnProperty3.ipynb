```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import os
import seaborn as sns
import pandas as pd
from scipy.io import loadmat

load_data = loadmat("property3.mat")
times = load_data["times"][0]
w2_results = load_data["w2_results"][0]
lu_results = load_data["lu_results"][0]
print(times)
print(w2_results)
print(lu_results)

plt.rcParams['axes.linewidth'] = 1  # 
plt.rcParams['figure.dpi'] = 300
plt.rc('font',family='Times New Roman')

figsize=(12, 7.2)
tick_fontsize = 30
label_fontsize  = 35

fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(111)

tick_fontsize=30
label_fontsize  = 30
plt.xlim((-0.1, 10.1))

x_ticks = [0, 2,  4, 6, 8, 10]
x_labels = [0, 2,  4, 6, 8, 10]
plt.xticks(x_ticks, x_labels, fontsize=tick_fontsize)

plt.ylim((-0.79, -0.74))
y_ticks =  [-0.79, -0.78, -0.77, -0.76, -0.75, -0.74]
y_labels = [-0.79, -0.78, -0.77, -0.76, -0.75, -0.74]
plt.yticks(y_ticks, y_labels, fontsize=tick_fontsize)
plt.gca().yaxis.grid(color='grey', linestyle='-', linewidth=1, alpha=0.3, zorder=0)
plt.gca().xaxis.grid(color='grey', linestyle='-', linewidth=1, alpha=0.3, zorder=0)

ln1 =ax.plot(times, w2_results*(-1.0), 'P', color='blue', linewidth=2, markersize=12, linestyle="-.")

ax2 = ax.twinx()
ax2.set_ylim((3.47,3.52))
y_ticks = [3.47, 3.48, 3.49, 3.5, 3.51, 3.52]
y_labels = [3.47, 3.48, 3.49, 3.5, 3.51, 3.52]
ax2.set_yticklabels(y_ticks, fontsize=tick_fontsize)
ln2 = ax2.plot(times, lu_results*(-1.0), 'd', color='green', linewidth=2, markersize=12, linestyle="--")



lns = ln1 + ln2
ax.legend(lns, [ ''r'$-\mathcal{W}_\mathcal{2}$', ''r'$-\mathcal{L}_{\mathcal{U}}$'], loc=1, fontsize=tick_fontsize)

ax.set_xlabel("Times of Instance Clone", fontsize=label_fontsize)
ax.set_ylabel('$-\mathcal{W}_\mathcal{2}$', fontsize=label_fontsize)
ax2.set_ylabel('$-\mathcal{L}_{\mathcal{U}}$', fontsize=label_fontsize)

plt.savefig('AnalysisOnICCProperty.pdf', bbox_inches='tight')
```
